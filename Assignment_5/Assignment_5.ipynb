# **Identifying Semantic Relationships using WordNet**

**Step 1: Install & Import Libraries**

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.corpus import wordnet as wn
```

**Step 2: Sample Text**

```python
words = ["happy", "big", "car", "dog"]
```

**Step 3: Synonym Extraction using WordNet**

```python
def get_synonyms(word):
    synonyms = set()
    for syn in wn.synsets(word):
        for lemma in syn.lemmas():
            synonyms.add(lemma.name())
    return synonyms

for word in words:
    print(f"\nSynonyms of '{word}':")
    print(get_synonyms(word))
```

**Step 4: Antonym Extraction using WordNet**

```python
def get_antonyms(word):
    antonyms = set()
    for syn in wn.synsets(word):
        for lemma in syn.lemmas():
            if lemma.antonyms():
                antonyms.add(lemma.antonyms()[0].name())
    return antonyms

for word in words:
    print(f"\nAntonyms of '{word}':")
    print(get_antonyms(word))
```

**Step 5: Hypernym Extraction using WordNet**

```python
def get_hypernyms(word):
    hypernyms = set()
    for syn in wn.synsets(word):
        for hyper in syn.hypernyms():
            hypernyms.add(hyper.name().split('.')[0])
    return hypernyms

for word in words:
    print(f"\nHypernyms of '{word}':")
    print(get_hypernyms(word))
```

**Step 6: Apply to Text Data**

```python
text = "The happy dog sat inside the big car"

tokens = text.lower().split()

for token in tokens:
    print(f"\nWord: {token}")
    print("Synonyms :", get_synonyms(token))
    print("Antonyms :", get_antonyms(token))
    print("Hypernyms:", get_hypernyms(token))
```
